### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep='|', index_col='user_id')
users.head()
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby('occupation').mean()
```

```python
users.info()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
# state = "nice" if is_nice else "not nice"
users['gender_n'] = users['gender'].apply(lambda x: 1 if x == 'M' else 0)
users
```

```python
# Proposta di Mario
cols = ['occupation']
users_M = users.loc[users.gender=='M', cols].groupby('occupation').size()
users_M
```

```python
tot_maschietti_per_occ = users.groupby('occupation').gender_n.sum()
```

```python
tot_per_occ = users.occupation.value_counts()
```

```python
ratio_masc = (tot_maschietti_per_occ / tot_per_occ)*100
ratio_masc.sort_values(ascending=False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby('occupation').age.agg(['min', 'max'])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(['occupation', 'gender']).age.mean()
```
